```python
statell = []
for Fips in df['FIPS_Code'].values:
    
    if Fips > 999:
        tmp = int(str(Fips)[-3:])
        if tmp!= 0:
            statell.append(tmp)
        else:
            statell.append(np.nan)
        

    else:
        #print(geoid)
        statenom =  np.nan
        statell.append(statenom)
```

```python
df['CountyNo'] = statell
```

```python
df = df.dropna()
```

```python
statefipsdict = dict()
for Fips,State in zip(df['FIPS_Code'].values,df['State'].values):
    
    if Fips > 999:
        if Fips > 9999:
            tmp = int(str(Fips)[:2])
        else:
            tmp = int(str(Fips)[:1])
        if Fips not in statefipsdict:
            if tmp!= 0:
                statefipsdict[tmp] = State
            
```

```python
import json

# python dictionary with key value pairs

# create json object from dictionary
json = json.dumps(statefipsdict)

# open file for writing, "w" 
f = open("statefipsdict.json","w")

# write json object to file
f.write(json)

# close file
f.close()
```
