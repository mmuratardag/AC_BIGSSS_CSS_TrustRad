```python
import pandas as pd
import numpy as np
# unemployment, racial distribution, %foreign born, education by county No., urbanization
```

```python
df=  pd.read_csv('../df/DECENNIALPL2020.P1_data_with_overlays_2022-04-22T144623.csv')
```

```python
len(df['NAME'].unique())
```

```python
statell = []
countyll = []
for geoid in df['GEO_ID'].values:
    tmp = geoid.find('US') +2
    tmp = geoid[tmp:]
    if len(tmp) == 5 and tmp[0] !=0:
        #if tmp[0]!=0:
        #print(tmp[:3])
        countyll.append(int(tmp[-3:]))
        statell.append(int(tmp[:2]))
    #else:
     #       statell.append(int(tmp[:3]))
    else:
        #print(geoid)
        statenom =  np.nan
        countyll.append(statenom)
        statell.append(statenom)

```

```python

df['CountyNo'] =  countyll
df['StateNo'] =  statell

#newdf = pd.DataFrame(data={'CountyNo' : statell[1:]})
```

```python
 
import json
  
# Opening JSON file
f = open('statefipsdict.json')
  
# returns JSON object as 
# a dictionary
statefipsdict = json.load(f)
f.close()
```

```python
o=0
for i in df.iloc[0]:
    print(o,i)
    o+=1

RaceCategories =['NativeAmerican (AmericanIndianAlaskaNative, NativeHawaianPacific)',
'Asian (EastAsian,SouthAsian)',
'Black',' White',' Mixed', 'OtherUnknown']
Columns = [[6,8],7,5,4,2,9]
```

```python
dfcols = df.columns
racedist = dict()
racedist["CountyNo"]= countyll[1:-1]
racedist["StateNo"]= statell[1:-1]
for kk, col in enumerate(Columns):
    racedist[RaceCategories[kk]] = []
    print(col)
    print(RaceCategories[kk],df[dfcols[col]].iloc[0])
    if col != 2:
        for ii in range(1,len(df)-1):
            tmp = pd.to_numeric(df[dfcols[col]].iloc[ii]).sum()
            racedist[RaceCategories[kk]].append(tmp)
    else:
        for ii in range(1,len(df)-1):
            tmp = pd.to_numeric(df[dfcols[2]].iloc[ii])  - pd.to_numeric(df[dfcols[3]].iloc[ii]) 
            racedist[RaceCategories[kk]].append(tmp)


newdf = pd.DataFrame(data=racedist)

newdf['CountyNo'] = pd.to_numeric(newdf['CountyNo'])

newdf['StateNo'] = pd.to_numeric(newdf['StateNo'])
newdf = newdf[['StateNo','CountyNo','NativeAmerican (AmericanIndianAlaskaNative, NativeHawaianPacific)',
       'Asian (EastAsian,SouthAsian)', 'Black', ' White', ' Mixed',
       'OtherUnknown']]
```

```python
newdf['StateNo'] = newdf['StateNo'].astype(str)
```

```python
newdf['STATE'] =newdf['StateNo'].replace(statefipsdict)
```

```python
def HerfindalIndex(vals):
    #print(vals)
    sumvals = sum(vals)
    #print(sumvals)
    vals= [val/sumvals for val in vals]
    tmp = 0
    for val in vals:
        tmp += val**2
    #print(tmp)
    return tmp

    
```

```python
newdf.columns[2:8]
```

```python
hhi = []
for i in range(len(newdf)):
    hhi.append(HerfindalIndex(newdf[newdf.columns[2:8]].iloc[i].values))
newdf['HHI'] = hhi
```

```python
newdf = newdf.drop('StateNo',axis=1)
```
